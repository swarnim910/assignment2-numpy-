```python
import matplotlib.pyplot as plt
```

```python
### TASK 1
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.plot(x, y, marker='o', linestyle='-', color='r', label="Plotting Points")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Axis Labelling")
plt.legend()
plt.show()
```

```python
### TASK 2
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.bar(students, marks, color=['blue','green','red','purple'])
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Performance Report")
plt.show()
```

```python
### TASK 3
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
colors = ['red','green','blue','yellow']

plt.pie(revenue, labels = regions, autopct='%1.1f%%', colors=colors)
plt.title("Revenue Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
random_integers = np.random.randint(1,100,1000)

plt.hist(random_integers, bins=10, color='red', edgecolor='black')

plt.xlabel("Integers")
plt.ylabel("Distribution")
plt.title("Frequency Distribution")
plt.show()
```
