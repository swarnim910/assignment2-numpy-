```python
import pandas as pd
```

```python
### TASK 1 
```

```python
data = {
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
  }
df = pd.DataFrame(data)
```

```python
df.head(5)
```

```python
df[['Age','Salary']].describe()
```

```python
hr_avg_salary = df[df['Department'] == 'HR']['Salary'].mean()
print(hr_avg_salary)
```

```python
### TASK 2
```

```python
df['Bonus'] = df['Salary'] * 0.10
print(df)
```

```python
### TASK 3
```

```python
filtered_df = df[(df['Age'] >=25) & (df['Age'] <=30)]
print(filtered_df)
```

```python
### TASK 4
```

```python
average_salary = df.groupby('Department')['Salary'].mean().reset_index()
print(average_salary)
```

```python
### TASK 5
```

```python
sorted_df = df.sort_values(by='Salary',ascending=True).reset_index()
print(sorted_df)
```

```python
sorted_df.to_csv('sorted_employees.csv',index=False)
```
